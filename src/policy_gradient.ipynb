```python
from environment import GridWorld
from agent import Agent
import matplotlib.pyplot as plt
```

```python
env=GridWorld()
```

```python
# For sigma = 0.1
agent = Agent(env, 0.1)
mean_and_standard_deviation_per_trial, mean_and_standard_deviation_per_episode = agent.hill_search()
```

```python
# Plotting the mean and standard deviation per trial
plt.figure(figsize=(20, 6))
plt.plot(mean_and_standard_deviation_per_trial[0], label='Mean')
plt.plot(mean_and_standard_deviation_per_trial[1], label='Standard Deviation')
plt.xlabel('Trials')
plt.ylabel('Mean and Standard Deviation')
plt.title('Mean and Standard Deviation per trial for sigma = 0.1')
plt.legend()
```

```python
# Plotting the mean and standard deviation per episode
plt.figure(figsize=(20, 6))
plt.plot(mean_and_standard_deviation_per_episode[0], label='Mean')
plt.plot(mean_and_standard_deviation_per_episode[1], label='Standard Deviation')
plt.xticks(range(0, len(mean_and_standard_deviation_per_episode[0])))
plt.xlabel('Episodes')
plt.ylabel('Mean and Standard Deviation')
plt.title('Mean and Standard Deviation per episode for sigma = 0.1')
plt.legend()
```

```python
# Plot error bars for the mean and standard deviation per trial
plt.figure(figsize=(20, 6))
plt.errorbar(range(0, len(mean_and_standard_deviation_per_trial[0])), mean_and_standard_deviation_per_trial[0], yerr=mean_and_standard_deviation_per_trial[1], fmt='.')
plt.xlabel('Trials')
plt.ylabel('Mean')
plt.title('Mean and Standard Deviation per trial for sigma = 0.1')
```

```python
# Plot error bars for the mean and standard deviation per episode
plt.figure(figsize=(20, 6))
plt.errorbar(range(0, len(mean_and_standard_deviation_per_episode[0])), mean_and_standard_deviation_per_episode[0], yerr=mean_and_standard_deviation_per_episode[1], fmt='o')
plt.xlabel('Episodes')
plt.ylabel('Mean')
plt.title('Mean and Standard Deviation per episode for sigma = 0.1')
```

```python
from environment import GridWorld
from agent import Agent
import plotly.graph_objects as go
```

```python
env=GridWorld()
fig = go.Figure()
for sigma in [0.1, 0.5, 0.9]:
    agent = Agent(env, sigma)
    average_gains = agent.hill_search()
    fig.add_trace(go.Scatter(x=list(range(len(average_gains))), y=average_gains, mode='lines', name=f'sigma={sigma}'))
```

```python
fig.show()
```
